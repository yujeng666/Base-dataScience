```python
import matplotlib.pyplot as plt
```

```python
#cara menamilkan grafik BAR
fig =plt.figure()
ax = fig.add_axes([0,0,1,1])
jurusan = ['RPL', 'PH', 'OTKP']
siswa = [90, 120, 130]
c =['yellow', 'red','black']
plt.title("PPDB SMK AL-MAADANI GARUT")
plt.bar(jurusan, siswa, color = c)
plt.show()
```

```python
#pie chart
fig = plt.figure()
ax = fig.add_axes([0, 0, 1,1])
ax.axis('equal')
jurusan = ['RPL', 'PH', 'OTKP']
siswa = [90, 120, 130]

ax.pie(siswa,labels=jurusan,autopct='%1.2f%%',)
plt.title("Data PPDB siswa al-madani")

plt.show()
```
